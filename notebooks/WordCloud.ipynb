## Importing Python Libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt; plt.style.use('seaborn')
from wordcloud import WordCloud, STOPWORDS
```

## 01 Reading Data

```python
# Reading Local Files
with open(file_path, encoding="utf-8") as fle:
  content = fle.read()
print("File data sample :", content[:20])
```

## 02 Preparing Data

```python
# Create stopword list:
stopwords = set(STOPWORDS)
```

```python
#Generate wordcloud data
wordcloud = WordCloud(stopwords=stopwords, max_words=25, \
                      background_color="white").generate(content)
```

## 03 Displaying the Wordcloud

```python
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

## 04 Enhancing Wordcloud

```python
#Add more words to ignore
stopwords.update(["may","see", "make", "time","every","us","must", "let", "much", "one"])

#Redo stop words. Limit number of words
wordcloud = WordCloud(stopwords=stopwords, max_words=20, \
                      background_color="azure").generate(content)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
