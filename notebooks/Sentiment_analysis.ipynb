```python
# Step 1: Import necessary libraries
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Ensure you have the VADER lexicon downloaded
nltk.download('vader_lexicon')

# Step 2: Initialize the SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Step 3: Read the file containing movie reviews
file_path = "/content/sample_data/Movie-Reviews.txt"

# Open and read the file
with open(file_path, 'r', encoding='utf-8') as file:
    reviews = file.readlines()

# remove empty lines
reviews = [review for review in reviews if review.strip() != ""]
print(len(reviews))

# Step 4: Analyze each review's sentiment
for review in reviews:
    review = review.strip()  # Remove leading/trailing whitespace
    sentiment_score = sia.polarity_scores(review)  # Get sentiment scores
    print(f"Review: {review}")
    print(f"Sentiment Score: {sentiment_score}\n")

# Optionally, categorize reviews into positive, negative, or neutral
positive_reviews = []
negative_reviews = []
neutral_reviews = []

for review in reviews:
    review = review.strip()  # Remove leading/trailing whitespace
    if review:  # Check if the review is not empty
        sentiment_score = sia.polarity_scores(review)  # Get sentiment scores
        compound_score = sentiment_score['compound']  # Use the compound score for classification

        # Categorize based on compound score
        if compound_score >= 0.5:
            positive_reviews.append(review)
        elif compound_score <= -0.5:
            negative_reviews.append(review)
        else:
            neutral_reviews.append(review)

        # Print the sentiment analysis results for each review
        print(f"Review: {review}")
        print(f"Sentiment Score: {sentiment_score}\n")

print("Positive Reviews:")
print(positive_reviews)
print("\nNegative Reviews:")
print(negative_reviews)
print("\nNeutral Reviews:")
print(neutral_reviews)
```

```python
# Count the number of reviews in each category
num_positive = len(positive_reviews)
num_negative = len(negative_reviews)
num_neutral = len(neutral_reviews)

print(f"Number of Positive Reviews: {num_positive}")
print(f"Number of Negative Reviews: {num_negative}")
print(f"Number of Neutral Reviews: {num_neutral}")
```

```python
import matplotlib.pyplot as plt

# Bar chart visualization
sentiment_counts = [num_positive, num_negative, num_neutral]
sentiment_labels = ['Positive', 'Negative', 'Neutral']

plt.figure(figsize=(8, 5))
plt.bar(sentiment_labels, sentiment_counts, color=['green', 'red', 'blue'])
plt.title('Sentiment Analysis of Movie Reviews')
plt.ylabel('Number of Reviews')
plt.show();
```

```python
# List to hold compound sentiment scores for each review
compound_scores = []

# Perform sentiment analysis and collect compound scores
for review in reviews:
    sentiment_score = sia.polarity_scores(review)
    compound_scores.append(sentiment_score['compound'])
```

```python
# Set up the figure
plt.figure(figsize=(10, 6))

# Create a bar plot
plt.bar(range(len(compound_scores)), compound_scores, color='blue')

# Adding labels and title
plt.axhline(y=0, color='gray', linestyle='--')  # Line to distinguish positive and negative sentiments
plt.title('Sentiment Analysis of Each Review')
plt.xlabel('Review Index')
plt.ylabel('Sentiment Score (Compound)')
plt.xticks(range(len(reviews)), [f'Review {i+1}' for i in range(len(reviews))], rotation=45)
plt.grid(True)

# Show the plot
plt.show()
```
