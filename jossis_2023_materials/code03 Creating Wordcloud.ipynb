```python
path='/content/drive/My Drive/14-Teaching/JoSSIS/data/'
filename = 'CourseDescription.txt'
file = path+filename
```

## Importing Python Libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt; plt.style.use('seaborn')
from wordcloud import WordCloud, STOPWORDS
```

## 01 Reading Data

```python
# Reading Local Files
f = open(file, encoding="utf-8")
filedata = f.read()
print("File data sample :", filedata[:4])
```

## 02 Preparing Data

```python
# Create stopword list:
stopwords = set(STOPWORDS)
```

```python
#Generate wordcloud data
wordcloud = WordCloud(stopwords=stopwords, max_words=25, \
                      background_color="white").generate(filedata)
```

## 03 Displaying the Wordcloud

```python
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

## 04 Enhancing Wordcloud

```python
#Add more words to ignore
stopwords.update(["many","using", "want", "value","play","build","daily"])

#Redo stop words. Limit number of words
wordcloud = WordCloud(stopwords=stopwords, max_words=10, \
                      background_color="azure").generate(filedata)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
