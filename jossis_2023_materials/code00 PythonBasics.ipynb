## List Comprehension

list comprehension structure

`[ expression for item in list if condition ]`




List comprehensions are Pythonâ€™s way of implementing the notation for sets as used in mathematics.

$$\{ x^2 : x \in [1,2,3,4,5] \; \text{if} \; x>2 \}$$

```python
squares = [x**2 for x in [1, 2, 3, 4, 5] if x > 2]
print(squares)
```

##  Functions

A simple function

```python
def greet_user():
    '''Display a simple greeting.'''
    print("Hello!")
```

```python
greet_user()
```

Passing an argument

```python
def greet_user(username):
    '''Display a personalized greeting.'''
    print("Hello, "+username+"!")
```

```python
greet_user("Jesse")
```

Returning a value

```python
def add_numbers(x, y):
    '''Add two numbers and return the sum'''
    return x + y
```

```python
sum = add_numbers(3,5)
print(sum)
```
