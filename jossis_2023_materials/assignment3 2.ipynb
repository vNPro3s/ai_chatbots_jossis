```python
import json
import pandas as pd  # Import pandas
from tqdm.notebook import tqdm  # Import tqdm for the progress bar

# Initialize list to store JSON dictionaries along with labels
json_list = []

# Use tqdm to create a progress bar for the DataFrame iteration
for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    label = row['Label']
    tweet = row['Tweet']
    aResults_string = doAnalyse(tweet)

    try:
        aResults = json.loads(aResults_string)
    except json.JSONDecodeError:
        print(f"Failed to decode JSON from doAnalyse result: {aResults_string}")
        continue  # Skip to the next iteration

    # Add label to JSON result
    aResults['Label'] = label
    # Add original Tweet to JSON result
    aResults['Tweet'] = tweet

    # Add the extended JSON result to list
    json_list.append(aResults)

# Create a new DataFrame from the list of extended dictionaries
new_df = pd.DataFrame(json_list)
```
